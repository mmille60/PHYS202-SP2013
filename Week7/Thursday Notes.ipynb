```python
stepsize=.001
y0 =10 
t = np.arange(0,1+stepsize,stepsize)
a=-9.81
dydt =lambda t,y: a*t
y1=np.zeros(len(t))
y1[0]=y0
for i in range(1,len(t)):
    y1[i]=y1[i-1]+dydt(t[i-1],y1[i-1])*stepsize
#print y1
#print t
plot(t,y1,label="euler method")
xlabel("time(s)")
ylabel("height")
y=y0+.5*a*t**2
plot(t,y,"r--",label="anylitical")
legend(loc="best")
show()
```
