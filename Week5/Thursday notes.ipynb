```python
f = lambda x: x**(0.5) +sin(x)
xmin =1.7
xmax=12.4

#find the maxium value of f within the boundary set by  ymax
subx =x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin= 0.
N=1E5
samples_x = xmin +(xmax-xmin)*random_sample(N);
value = 0
for i in samples_x:
    value+=f(i)
IN=(xmax-xmin)*(1/N)*(value)
print IN
```

```python
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)
c = np.ones(x.size)
plot(x,(1/N)*value*c)
show()
```

```python
Iana = lambda x:(2./3.)*x**(3./2.) -cos(x)
Iexact = Iana(xmax) -Iana(xmin)
Ierr = 100*(Iexact-IN)/Iexact
print "I_exact",Iexact
print "I_histormiss =",IN
print "Errot on I = ",Ierr,"%"
```

This sample method returned a aproximatly similar error to the hit and miss method however it required fewer samples to complete
