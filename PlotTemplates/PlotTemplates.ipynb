Simple graphing template for matplotlib functions and data plots

Line plots
--

```python
#creating data array
y = np.arange(-5.0,5.0+0.1,.001)#creating an array of values from -5 to 5 
E = cosh(y)#maxing an array of values for cosh can change this function
p=sinh(y)#maxing an array of values for sinh can change this function
```

```python
#customization
plot(y,E, color= "blue",linewidth=2.5,linestyle="--",label="$function 1$")#plots y vs E can change line look here
plot(y,p, color = "red",linewidth=2.5,linestyle=":",label="$function 2$")#plots y vs p can chancge line look here
xlabel("$x label$")#input x lable here
ylabel("$y label$")#input y lable here
#Add tittle for convience
title("A simple plot",fontsize=20)
grid(True)#can turn grid lines on or off
legend(loc="lower left")#can place ledgend
ax = gca()
ax.spines['right'].set_color('none') # bellow you can turn the spines on or off
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color ="black")#turns on horizontal axis at y=0
axvline(x=0, color="black")#turns on vertical axis at x=0
show()
```

subplots

```python
subplot(2,1,1)#can change subplot value to change placing of graphs
plot(y,E, color= "blue",linewidth=2.5,linestyle="--",label="$function 1$")
xlabel("$x label$")#input x lable here
ylabel("$y label$")#input y lable here
#Add tittle for convience
title("A simple sub plot",fontsize=20)
grid(True)#can turn grid lines on or off
legend(loc="lower left")#can place ledgend
ax = gca()
ax.spines['right'].set_color('none') # bellow you can turn the spines on or off
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color ="black")#turns on horizontal axis at y=0
axvline(x=0, color="black")#turns on vertical axis at x=0
subplot(2,1,2)
plot(y,p, color = "red",linewidth=2.5,linestyle=":",label="$function 2$")#plots y vs p can chancge line look here
xlabel("$x label$")#input x lable here
ylabel("$y label$")#input y lable here
#Add tittle for convience

grid(True)#can turn grid lines on or off
legend(loc="upper left")#can place ledgend change placing depending on plot
ax = gca()
ax.spines['right'].set_color('none') # bellow you can turn the spines on or off
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axhline(y=0, color ="black")#turns on horizontal axis at y=0
axvline(x=0, color="black")#turns on vertical axis at x=0
show()
```
