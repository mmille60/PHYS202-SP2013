```python
k=8.987*10**9#is in the week 4 file
def pointFeild(x,y,q,Xq,Yq):
        """put in x,y,q,Xq,Yq returns a tuple of the electric field componets Ex , Ey"""
        Ex=(k*q)*(x-Xq)/((x-Xq)**2+(y-Yq)**2)**.5
        Ey=(k*q)*(y-Yq)/((x-Xq)**2+(y-Yq)**2)**.5
        return (Ex,Ey)
```
