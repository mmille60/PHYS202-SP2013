```python
import numpy as np
x = arange(-4,4,.01)
y = 1/cosh(x)
U = -y**2
i=0+1j
l=[0,.5,1,2]
line=['r-','r--','r-.','r:']
m=0
for k in l:
    psik= ((i*k-tanh(x))/(i*k+1)*exp(i*k*x))
    plot(x,psik*psik.conjugate(),line[m],linewidth=2.5,label="$\psi^2,k=%s$"%k)
    m+=1
plot(x,U,linewidth=2.5,label="$U(x)$")
xlabel("$x(rad)$",fontsize=15)
ylabel("$U(x), \psi^2$",fontsize=15)
grid(True)
xticks(arange(-pi,pi+1,pi/2),["$-{\pi}$", "$0$", "$+{\pi}/2$","$+{\pi}$"])
legend(loc='lower left')
show()
```

```python
### Question 3
data = loadtxt("sunspots.txt",float)
sunx = data[:,0]
suny = data[:,1]
sunx_1000=data[:1001, 0]
suny_1000=data[:1001,1]
plot(sunx_1000,suny_1000,color = "red",linewidth=2.5,linestyle="--",label="$cos(4{\pi}t)$")
def averag(array):
    total=0
    count=1
    for i in array:
        total+=i
        count+=1
    avg =(total)/count
    return avg
avgs = arange(1001)
count = 0
for i in range(0,1001):
    avgs[count] = averag(suny_1000[0:i])
    count +=1
plot(sunx_1000,avgs,color= "blue",linewidth=2.5,linestyle=":",label="$sin(4{\pi}t)$")
title("A simple plot",fontsize=20)
grid(True)
legend(loc="upper right")
show()
```

```python
data = loadtxt("stars.txt",float)
x= data[:,0]
y= data[:,1]
scatter(x,y,s=5,color="green",label="$Magnitude$")
title("Hertzberg/Russel Diagram",size=18)
ylabel("Magnitude",size=12)
xlabel("Temperature(K)",size=14)
xlim(xmax=0, xmin=16000)
ylim(ymax=-5,ymin=20)
annotate("Main Sequence" , xy=(6000,5),xytext=(14000,5),arrowprops=dict(facecolor='black',shrink=.08),)
annotate('White Dwarfs', xy=(7200,14),xytext =(15000,15),arrowprops=dict(facecolor='black',shrink=0.08),)
annotate('Red Giants', xy=(4700,0),xytext=(10000,-2),arrowprops=dict(facecolor='black',shrink=0.08),)
grid(True)
legend(loc="lower center")

show()
```
