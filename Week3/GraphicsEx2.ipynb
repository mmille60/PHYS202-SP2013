```python
import numpy as np
x = arange(-4,4,.01)
y = 1/cosh(x)
U = -y**2
i=0+1j
krange=[0,.5,1,2]
l=['r-','r--','r-.','r:']
t=0
for k in krange:
    psik= ((i*k-tanh(x))/(i*k+1)*exp(i*k*x))
    plot(x,psik*psik.conjugate(),l[t],linewidth=2.5,label="$\psi^2,k=%s$"%k)
    t+=1
plot(x,U,linewidth=2.5,label="$U(x)$")
xlabel("$x(rad)$",fontsize=15)
ylabel("$U(x), \psi^2$",fontsize=15)
grid(True)
xticks(arange(-pi,pi+1,pi/2),["$-{\pi}$", "$0$", "$+{\pi}/2$","$+{\pi}$"])
legend(loc ="lower left",prop={'size':6})
show()
```

```python
### Question 3
data = loadtxt("sunspots.txt",float)
sunx = data[:,0]
suny = data[:,1]
sunx_1000=data[:1001, 0]#Here you can change how many data point you would like
suny_1000=data[:1001,1]
plot(sunx_1000,suny_1000,color = "red",linewidth=2.5,linestyle="--",label="$Sunspots$")
def averag(array):
    total=0
    count=1
    for i in array:
        total+=i
        count+=1
    avg =(total)/count
    return avg
avgs = arange(1001)
count = 0
for i in range(0,1001):#note to change how far you want the average to got change here
    avgs[count] = averag(suny_1000[0:i])
    count +=1
plot(sunx_1000,avgs,color= "blue",linewidth=2.5,linestyle=":",label="$running average$")
title("Sunspots vs time",fontsize=20)
grid(True)
legend(loc="upper right")
xticks(arange(0,1001,1000/6),["$1749$", "$1763$", "$1777$","$1791$","$1805$", "$1819$", "$1832$"])
xlabel("Time(years)",size=14)
ylabel("sunspots",size=14)
show()
```
