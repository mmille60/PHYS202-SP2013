```python
t=arange(-6,0,.01)
f_x=exp(t)
t_1=arange(0,6,.01)
f_x2=exp(-t_1)
plot(t,f_x)
plot(t_1,f_x2, color='blue')
title("Plot of f(x) in natural units")
ylabel('$increments\ of\ C$',size=12)
xlabel('$\\frac{1}{a}\ units\ of\ m^{-1}$',size=12)
grid(True)
show()
```

```python
t=arange(-8,8,.01)
f_x3=1/(1+t*t)
plot(t,f_x3)
title("Plot of A(K) in natural units")
xlabel(r'$\frac{1}{k^2} \ (m^{2})$',size=12)
ylabel(r'$\frac{Ca}{\pi} \ (m^{-1})$' ,size=12)
grid(True)
show()
```
