creat a very simple plot

```python
import numpy as np
t = np.arange(0.0,1+0.1,.01)
c = np.cos(2*2*pi*t)
s = np.sin(2*2*pi*t)
```

```python
plot(t,s, color= "blue",linewidth=2.5,linestyle="--",label="$sin(4{\pi}t)$")
plot(t,c, color = "red",linewidth=2.5,linestyle=":",label="$cos(4{\pi}t)$")
xlabel("time(s)")
ylabel("Voltage(mV)")
#Add tittle for convience
title("A simple plot",fontsize=20)
grid(True)
legend(loc="lower left")
savefig("simple_plot.pdf")
```
