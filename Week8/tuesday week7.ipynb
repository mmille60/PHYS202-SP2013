```python
x = arange(0.,pi+0.001,0.01)
f=lambda x:sin(x)/2.
F=lambda x: (cos(0)-cos(x))/2.
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob prob is an array, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data"""
    z=numpy.random.random_sample(N);
    C=cumsum(prob)
    
    a=sum(prob)
    C=C/a
    T=C.searchsorted(z)
    new=x[T]
    return new
```
