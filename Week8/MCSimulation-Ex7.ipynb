```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob prob is an array, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data"""
    z=numpy.random.random_sample(N);
    C=cumsum(prob)
    
    a=sum(prob)
    C=C/a
    T=C.searchsorted(z)
    new=x[T]
    return new
```
