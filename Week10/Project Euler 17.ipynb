If the numbers 1 to 5 are written out in words: one, two,three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
basics=["zero","one","two","three","four","five","six","seven","eight","nine","ten","eleven","twelve","thirteen","fourteen","fifteen","sixteen","seventeen","eighteen","nineteen","twenty"]
by10ns=["zero","ten","twenty","thirty","forty","fifty","sixty","seventy","eighty","ninety"]
```

```python
#this one is for testing
#creating a function that returns the str legth of a numbers name and that name
def number_to_leters_to_length(num):#only neeeds to work from 1 to 1000
    if num%1000==0:#handels 1000 canse
        s="onethousand"
    elif num%100==0:#for multipuls of 100
        first=num/100
        s=basics[first]+"hundred"
    elif num%100!=0:#for all other numbers
        if num/100.>1:#is the number greater than 100 or less than if greater enter this loop
            first=num/100
            if num%100<10:#101-110
                secound="and"+basics[num%100]
                s=basics[first]+"hundred"+secound
            elif num%100<21 and num%100>9:#111-120
                secound="and"+basics[num%100] 
                s=basics[first]+"hundred"+secound
            elif num%100>20 and num%10==0:#greater than 100 but multipuls of 10
                secound="and"+by10ns[num%100/10]
                s=basics[first]+"hundred"+secound
            else:#everything else
                secound="and"+by10ns[num%100/10]+basics[num%100%10]
                s=basics[first]+"hundred"+secound
        elif num/100.<1:#if number less than 100 enter here
            if num>20 and num%10!=0:
                s=by10ns[num/10]+basics[num%10]
            elif num>20 and num%10==0:
                s=by10ns[num/10]
            else:
                s=basics[num]
    return len(s),s
```
