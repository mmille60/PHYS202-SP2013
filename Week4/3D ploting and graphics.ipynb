Electric Dipole Potential
===

$V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
#investigating mesh grid
dx=0.06
dy=0.06
x,y=meshgrid(arange(-2.,2.,dx),arange(-2,2,dy))
print x.ndim
print y.shape
print y
print x
```

```python
k=1
q=1
d=1
Vxy=((k*q)/(x**2+(y-d)**2)**.5)-((k*q)/(x**2+(y+d)**2)**.5)
print Vxy
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p=ax.plot_surface(x,y,Vxy,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)#use tab next to cm. to find out other colors

cd=fig.colorbar(p, shrink=0.5)
```

```python
levels=array([8,4,2,1,.5,.25,0,-.25,-0.5,-1,-2,-4,-8])
contour(x,y,Vxy,levels,linewidths=4,cmap=cm.coolwarm)
```

```python
k=8.8987*(10**9)
def pointPotential(x,y,q,posx,posy):
    """Vxy is the eletra potential it takes x,y,q,posx,posy and returns the potential"""
    Vxy=((k*q)/((x-posx)**2+(y-posy)**2)**.5)
    return Vxy
```

```python
x = 1.; y = 1.
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y = meshgrid(arange(-4.,4.,0.06),arange(-4.,4.,0.06))
Vxy = pointPotential(x,y,1e-9,0.,0.)
print Vxy
```

```python

def dipolePotential(x,y,q,d):
    Vxy=pointPotential(x,y,q,(d/2.),0)+pointPotential(x,y,-q,(-d/2.),0)
    return Vxy
def dipolePotentialy(x,y,q,d):
    Vy=pointPotential(x,y,q,(0),(d/2.))+pointPotential(x,y,-q,(0),(-d/2.))
    return Vy
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p=ax.plot_surface(x,y,dipolePotential(x,y,exp(-9),4),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p=ax.plot_surface(x,y,dipolePotential(x,y,exp(-9),4)+dipolePotentialy(x,y,exp(-9),4),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```

```python
k=8.8987*(10**-9)
def pointPotential(x,y,q,posx,posy):
    """Vxy is the eletra potential it takes x,y,q,posx,posy and returns the potential"""
    Vx=((k*q)/((x-posx)**2+(y-posy)**2)**.5)
    return Vx
def dipolePotentialangle(x,y,q,d,Theta):
    """returns dispole pontial betweentwo particles seperated by distance d and angle theta with axis"""
    Vxy= pointPotential(x,y,q,(d/2.*cos(Theta)),(d/2.*sin(Theta)))+pointPotential(x,y,-q,-(d/2.*cos(Theta)),(-d/2.*sin(Theta)))
    return Vxy
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p=ax.plot_surface(x,y,dipolePotentialangle(x,y,exp(-9),4,30),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```
