```python
#creating a function to take a derivative
def finiteDifference(x,y):
    """returns the derivative of y with respect to x forward differentiation"""
    import numpy as np
    dy = diff(y) #y(n+1)-y(n)
    dx = diff(x)
    dydx1 =zeros(y.shape,float)
    dydx1[:-1] = dy/dx
    dydx1[-1] = (y[-1] -y[-2])/(x[-1] - x[-2])#diferintionation method
    return dydx1
```

```python
#created a secound difrentionator for fun
def center_Difference(x,y):
    """performs center differentiation"""
    dy = diff(y) #y(n+1)-y(n)
    dx = diff(x)
    dydx2 =zeros(y.shape,float)
    dydx2[1:-1] = (y[2:] -y[:-2])/(x[2:]- x[:-2])
    dydx2[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx2[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx2
```

Problem 3 part b

```python
#conparing diffrent methods of diferenteation 
x = arange(0,10.01,0.1)
y= x**5
z=finiteDifference(x,y)
print z
f=center_Difference(x,y)
print f
AnalyticalDifrentioator=5*x**4
print AnalyticalDifrentioator
plot(x,z,"r.",label="$forward$ $differentiation$")
plot(x,f,"b.",label="$center$ $differentiation$")
plot(x,AnalyticalDifrentioator,"g.",label="$analytical$ $differentiation$")
legend(loc="upper left")
xlabel("x values")
ylabel("y values")
show()
```

When both diffrenteation methods are compared to the anylitical method it can be seen that both methods do a good joob of takeing the drivative. The shows my work was done corectly.

```python
#Plots the Errors of the methods of taking derivatives
x = arange(0,10.01,0.001)
y= x**5
z=finiteDifference(x,y)
f=center_Difference(x,y)
dydxExact=5*x**4
percentError1 = 100*abs(z[1:] - dydxExact[1:])/dydxExact[1:]#calculates percent error
percentError2 =100*abs(f[1:] - dydxExact[1:])/dydxExact[1:]
semilogy(x[1:],percentError2,label="center differentiation")#plots the errors
semilogy(x[1:],percentError1,label="forward differentiation")
ylabel("percent error compared dydx exact")
xlabel("range of x values")
legend(loc="upper right")
title("Error in differentiation")
show()
```

Accuracy of methods of differentiation ploted above

```python
#creating the function
def foutPtFiniteDiff(x,y):
    """takes an array x and a function y as inputs and returns the derivative of the function y"""
    h= x[2]-x[1]#sets the splacing
    dydx =zeros(y.shape,float)#initializes the matrix
    for i in range(2,len(y)-2):#runs through all the values
        dydx[i]=(y[i-2]-8*y[i-1]+8*y[i+1]-y[i+2])/(12*h)
    return dydx
```

Problem 4 part b testing
--

```python
x = arange(0,10.01,0.001)
y= x**5#function to take derivative of
r=foutPtFiniteDiff(x,y)
plot(x,r,linestyle =":",label="$four point$ $differentiation$")
dydxExact=5*x**4
plot(x,dydxExact,linestyle ="--",label="$analytical$ $differentiation$")
legend(loc="upper left")
xlabel("x values")
ylabel("derivatives of $x^5$")
show()
```

```python
x = arange(0,10.01,0.001)
dydxExact=5*x**4
percentError3 =100*abs(r[2:] - dydxExact[2:])/dydxExact[2:]#finding the error between the anaylitical solution and the four point meth
semilogy(x[2:],percentError3,label="forward differentiation error")
ylabel("percent error compared dydx exact")
xlabel("range of x values")
legend(loc="upper right")
title("Error in differentiation")
show()
```

The above plots show that the four point method is working corectly with a spacing of .001 although it fails at the last data points which is to be expected
