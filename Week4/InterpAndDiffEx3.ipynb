```python
import numpy as np
def finiteDifference(x,y):
    """returns the derivative of y with respect to x"""
    dy = diff(y) #y(n+1)-y(n)
    dx = diff(x)
    dydx1 =zeros(y.shape,float)
    dydx1[:-1] = dy/dx
    dydx1[-1] = (y[-1] -y[-2])/(x[-1] - x[-2])
    return dydx1
```

```python
def center_Difference(x,y):
    dy = diff(y) #y(n+1)-y(n)
    dx = diff(x)
    dydx2 =zeros(y.shape,float)
    dydx2[1:-1] = (y[2:] -y[:-2])/(x[2:]- x[:-2])
    dydx2[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx2[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx2
```

```python
x = arange(0,10.01,0.001)
y= x**5
z=finiteDifference(x,y)
print z
f=center_Difference(x,y)
print f
```

```python
dydxExact=5*x**4
plot(x,z,color="r")
plot(x,dydxExact,linestyle ="--")
show()
#shows diffrence between numaric and analitical difretiation
```

```python
x = arange(0,10.01,0.001)
percentError1 = 100*abs(z[1:] - dydxExact[1:])/dydxExact[1:]
percentError2 =100*abs(f[1:] - dydxExact[1:])/dydxExact[1:]
semilogy(x[1:],percentError2,label="center difference")
semilogy(x[1:],percentError1,label="forward difference")
ylabel("percent error with dydx exact")
xlabel("range of x values")
legend(loc="upper right")
show()
```
