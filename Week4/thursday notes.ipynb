```python
dx=0.0006
dy=0.0006
x,y=meshgrid(arange(-.1,.1,dx),arange(-.1,.1,dy))
posx=0
posy=-5*10**(-2)
q1=-2*10**-9
q2=3*10**-9
d=1*10**-2
total=pointPotential(x,y,q1,posx,posy)+dipolePotential(x,y,q2,d)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,total,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
#p=ax.plot_surface(x,y,pointPotential(x,y,q1,posx,posy),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)#use tab next to cm. to find out other colors
#y=ax.plot_surface(x,y,dipolePotential(x,y,q2,d),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
show()
```

```python
levels=array([8,4,2,1,.5,.25,0,-.25,-0.5,-1,-2,-4,-8])
levels=levels*25
contour(x,y,total,levels,linewidths=4,cmap=cm.coolwarm)
show()
```

Using boolean masks to find data
==

```python
theta = arange(0.,2*pi,pi/100)
plot(theta,sin(theta))
show()
```

```python
mask=(sin(theta)>0) & (sin(theta)<.5)
print mask
```

```python
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
```

```python
y=sin(theta[mask])
jj= (y==max(y))
y[jj]=NaN
```

```python
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```
