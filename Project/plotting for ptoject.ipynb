```python
data = loadtxt("rpp2012.txt",str)
x = data[:,1]
y = data[:,4]
yee = data[:,5]
xee = data[:,7]
plab=zeros(len(x))#GEV/c
sig=zeros(len(y))#MB
plaberr=zeros(len(x))
sigerr=zeros(len(y))
for i in range(0,len(x)):
    plab[i]=float(x[i])
    sig[i]=float(y[i])
    plaberr[i]=float(xee[i])
    sigerr[i]=float(yee[i])
```
