```python
data = loadtxt("rpp2012.txt",str)
x = data[:,1]
y = data[:,4]
yee = data[:,5]
xee = data[:,7]
plab=zeros(len(x))#GEV/c
sig=zeros(len(y))#MB
plaberr=zeros(len(x))
sigerr=zeros(len(y))
for i in range(0,len(x)):
    plab[i]=float(x[i])
    sig[i]=float(y[i])
    plaberr[i]=float(xee[i])
    sigerr[i]=float(yee[i])
```

```python
data = loadtxt("rpp2012-pp_elastic.txt",str)
xe = data[:,1]
ye = data[:,4]
yeel = data[:,5]
xeel = data[:,7]
plab_elas=zeros(len(xe))#GEV/c
sig_elas=zeros(len(ye))#MB
elaplaberr=zeros(len(xe))
elasigerr=zeros(len(ye))
for i in range(0,len(xe)):
    plab_elas[i]=float(xe[i])
    sig_elas[i]=float(ye[i])
    elaplaberr[i]=float(xeel[i])
    elasigerr[i]=float(yeel[i])
from scipy.interpolate import interp1d
newsig = interp1d(plab_elas,sig_elas)(10**4)#interpulates temp and resistance values
print newsig
```

```python
ax = gca()
scatter(plab,sig,s=.8)
scatter(plab_elas,sig_elas,color="green",s=.8)
ax.set_xscale('log')
ax.set_yscale('log')
#loglog(plab,sig)
#loglog(plab_elas,sig_elas)
ax.scatter
ylim(0,350)
ylabel("Cross section(mb)",fontsize=18)
xlabel(r"$P_{lab}(\frac{Gev}{c})$",fontsize=18)
errorbar(plab,sig,xerr=plaberr,yerr=sigerr)
errorbar(plab_elas,sig_elas,xerr=elaplaberr,yerr=elasigerr)
ax.annotate("$total$",xy=(300,50),fontsize=14)
ax.annotate("$pp$",xy=(70,20),fontsize=14)
ax.annotate("$elastic$",xy=(200,10),fontsize=14)
ax.annotate(r"$P_{lab}(\frac{Gev}{c})$",xy=(10**7,7),fontsize=25)
grid(True)
show()
```
