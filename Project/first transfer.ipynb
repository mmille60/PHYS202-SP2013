```python
from scipy.optimize import curve_fit
def func(s,Z,B,Y,Y2,n1,n2):
    #hb=6.626*10**-34
    #c=3*10**8
    #n1=.462
    #n2=.550
    ma=1.67*10**-27
    mb=1.67*10**-27
    M=2.15
    sM=(ma+mb+M)**2
    sig=Z+B*(log(s/sM))**2+Y*(sM/s)**n1-Y2*(sM/s)**n2
    return sig
x=linspace(0,10**5,100)
popt, pcov = curve_fit(func, plab_elas, sig_elas)
print popt
#errorbar(s,y,yn-y,0,'r.') 
#plot(plab_elas,func(plab_elas,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5]))
s = linspace(2,100000.02,.01)
#plot(s,func(s,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5]))
show()


#Make an array with one more x value
xex = zeros(len(plab_elas)+1,float)
xex[0:-1] = plab_elas.copy()
for i in range(10**3,10**6):
    xex[-1] = i

#plot(xex,func(xex,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5]))
curvefit=func(xex,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5])
plot(xex,curvefit)
show()
```

```python
def find_interacting_distance(energy):
    from scipy.interpolate import interp1d
    crossection_total = interp1d(plab,sig)(energy)
    crossection_elastic = interp1d(xex,curvefit)(energy)
    crossection_inilastic= crossection_total-crossection_elastic
    d=((crossection_inilastic*10**-1)/pi)**.5
    return d
find_interacting_distance(10**1)
```

```python
def random_r(atom):
    N=number
    z=numpy.random.random_sample(Woodsaxson(atom)[3]);
    F1=cumsum(Woodsaxson(atom)[0])
    a=sum(Woodsaxson(atom)[0])
    F1=F1/a
    T=F1.searchsorted(z)
    r=Woodsaxson(atom)[1]
    radius_dis=r[T]
    theta=zeros(len(radius_dis))
    t=0
    for i in range(0,len(radius_dis)):
        theta[t]=random.uniform(0,2*pi)
        t+=1
    return (radius_dis,theta,Woodsaxson(atom)[0],Woodsaxson(atom)[1],Woodsaxson(atom)[2],Woodsaxson(atom)[3],Woodsaxson(atom)[4])
```
