Section 1 a) reading in data

total colition data

```python
#this part of the cose is to reaD in the data from the particle website
data = loadtxt("rpp2012.txt",str)#I read the proton data on total crossections
#here I take the data out that I want
x = data[:,1]#energy data
y = data[:,4]#cross section data
yee = data[:,5]#error
xee = data[:,7]#error
#pre-alocate 4 arrays
plab=zeros(len(x))#GEV/c
sig=zeros(len(y))#mb
plaberr=zeros(len(x))
sigerr=zeros(len(y))
for i in range(0,len(x)):#convert all the string data to floats
    plab[i]=float(x[i])
    sig[i]=float(y[i])
    plaberr[i]=float(xee[i])
    sigerr[i]=float(yee[i])
```

Section 1 b) curve fiting
----

I needed to fit elatic data so we can use graph at higher energies. I fit this curve to an equation from the particle group can be found on page 9 of the following link.

http://pdg.lbl.gov/2012/reviews/rpp2012-rev-cross-section-plots.pdf

```python
from scipy.optimize import curve_fit#import the curvefiting tool boxs
def func(s,Z,B,Y,Y2,n1,n2):#define the function to fit to
    #some constants to be used
    ma=1.67*10**-27
    mb=1.67*10**-27
    M=2.15
    sM=(ma+mb+M)**2
    #equantion bellow is what I fit the elastic data to
    sig=Z+B*(log(s/sM))**2+Y*(sM/s)**n1-Y2*(sM/s)**n2#
    return sig
popt, pcov = curve_fit(func, plab_elas, sig_elas)#here I fit the elastic data
print popt#print the curve fit parameters found



#Make an array with one more energy values to plot against
xex = zeros(len(plab_elas)+1,float)
xex[0:-1] = plab_elas.copy()#copy the elatic data
for i in range(10**3,10**6):#here I extend the data
    xex[-1] = i


curvefit=func(xex,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5])#obtain extend curve fit
plot(xex,curvefit)

show()
```

```python
#allows me to interpalate elastic data
#note interation distances are in fm
def find_crossection_elatic(energy):#note inergy in Gev/c
    from scipy.interpolate import interp1d#inport interpertation tools
    crossection_total = interp1d(plab,sig)(energy)#interpalate total data for that cross section
    crossection_elastic = interp1d(xex,curvefit)(energy)#interpalate elastic data for that cross section
    crossection_inilastic= crossection_total-crossection_elastic#the inilatic cross section will be the total minus the elatic
    d=((crossection_inilastic*10**-1)/pi)**.5#to find interaction distance
    return crossection_elastic
```

```python
#this funtion alows me to interpetlate the crossection data and extrac interaction distances in fm
#note interation distances are in fm
def find_interacting_distance(energy):
    """given energy in GeV/c returns crossection inelatic"""
    from scipy.interpolate import interp1d#inport interpertation tools
    crossection_total = interp1d(plab,sig)(energy)#interpalate total data for that cross section
    crossection_elastic = interp1d(xex,curvefit)(energy)#interpalate elastic data for that cross section
    crossection_inilastic= crossection_total-crossection_elastic#the inilatic cross section will be the total minus the elatic
    d=((crossection_inilastic*10**-1)/pi)**.5#to find interaction distance
    return d
find_interacting_distance(10**1)#exsample of function working and return a interation distance in fm
```

```python
import random
def Woodsaxson(atom):#need to input the atom you want the data on
    """given atom type returns woodsaxson equation along with other atom peramiters"""
    p_r=lambda r: (1+(w*r**2)/R0)/(1+exp((r-R0)/a))#this is the woodsaxson problity distrabution
    
    #each atom can be called by its abriviation or A number
    #all the if satements are just the parameters for different atoms
    #note all distances are in fm
    if atom=='O' or atom ==16:
        w=-0.051
        R0=2.608
        a=.513
        n=16
    elif atom=='C' or atom==12:
        w=0.
        R0=2.47
        a=0.
        n=12
    elif atom=='Al' or atom==27:
        R0=3.07
        a=.519
        w=0.
        n=27
    elif atom=='S' or atom==32:
        R0=3.458
        a=.61
        w=0
        n=32
    elif atom=="Ca" or atom==40:
        w=-0.161
        R0=3.76
        a=.586
        n=40
    elif atom=='Ni' or atom==59:
        R0=4.309
        a=.516
        w=-0.1308
        n=59
    elif atom=='Cu' or atom==64:
        R0=4.2
        a=.596
        w=0
        n=64
    elif atom=='W' or atom==184:
        R0=6.51
        a=.535
        w=0
        n=184
    elif atom=='Au' or atom==197:
        R0=6.38
        a=.535
        w=0
        n=197
    elif atom=='Pb' or atom==208:
        R0=6.68
        a=.546
        w=0
        n=208
    elif atom=='U' or atom==238:
        R0=6.68
        a=.6
        w=0
        n=238
    r=np.arange(0,1.1*R0,.01)#returns an array of radius values from 0-1.3 times the radius of the atom
    b=random.uniform(0, 1.5)*R0#creats a random interacting distance b
    return (p_r(r),r,b,n,R0)
```

random_r creats the nucleons for the colition by assigning theta and r values.

```python
def random_r(atom):#need to imput the atom you want the data on
    """given atom calls Woodsaxson to create a nucleus with peramiters acording to atom type"""
    #creats random nulei from the woodsaxson distribution using radomsampling methods 
    #to see how this exsacts function works please see Woodsaxson folder as it has plot examples 
    z=numpy.random.random_sample(Woodsaxson(atom)[3]); #take the corect number of samples n dependy on atom type
    F1=cumsum(Woodsaxson(atom)[0])#continuosly sums the problity distibution
    a=sum(Woodsaxson(atom)[0])#sums the problity array
    F1=F1/a#normalizes
    T=F1.searchsorted(z)#finds where the random sample fits in the problity ditribution
    r=Woodsaxson(atom)[1]#returns radius peramiters for the atom of intrest
    radius_dis=r[T]#asignes a radius value in fm to the radom sample acording to the problity distibution 
    theta=zeros(len(radius_dis))#intialize the theta aray
    t=0#tet my counter to 0
    for i in range(0,len(radius_dis)):
        theta[t]=random.uniform(0,2*pi)#generate random theta values from 0 to 2pi
        t+=1#increase counter
    return (radius_dis,theta,Woodsaxson(atom)[0],Woodsaxson(atom)[1],Woodsaxson(atom)[2],Woodsaxson(atom)[3],Woodsaxson(atom)[4])#return all needed info
```

```python
def collision_data_plot(atom1,atom2,Nc,energy):#need the two atoms the number of samples you want Nc and the energy you want the colitions to take place at
    """given atom 1 atom 2 number of runs and energy in GeV/c returns plots of collision data"""
    import numpy as np
    x=zeros(Nc)#intialize two arays the with leghts equal to the number of atoms you want to sample
    y=zeros(Nc)
    d=find_interacting_distance(energy)#gives radius of the proton at this energy
    for b in range(0,Nc):#the loop goes through the counting done in ploting_2 the number of times indicated by Nc
        count=0#sets count to 0
        atom1_dis=random_r(atom1)#gets the data on the nucleons
        atom2_dis=random_r(atom2)
        atom1_distance_cordinates=((atom1_dis[0]*cos(atom1_dis[1])),(atom1_dis[0]*sin(atom1_dis[1])))#converts to x,y cordinates
        atom2_distance_cordinates=(((atom2_dis[0]*cos(atom2_dis[1]))+atom1_dis[4]+atom2_dis[4]),(atom2_dis[0]*sin(atom2_dis[1])))
        #seperate x,y data
        x_atom1=atom1_distance_cordinates[0]
        y_atom1=atom1_distance_cordinates[1]
        x_atom2=atom2_distance_cordinates[0]
        y_atom2=atom2_distance_cordinates[1]
        count=0
        #the bellow for loops determine if the nuclei colide
        for i in range(0,len(x_atom1)):
            for c in range(0,len(x_atom2)):
                distance=((x_atom1[i]-x_atom2[c])**2+(y_atom1[i]-y_atom2[c])**2)**.5#finds distacne between each nucleon
                if distance<2*d:
                    count+=1

        x[b]=atom1_dis[4]+atom2_dis[4]#stors the distance between each atom in array
        y[b]=count#adds the count to an array
    figure(figsize=(15,15))#set the figure size
    subplot(2,1,2)#plots the lower graph in the subplot
    
    yforploting=np.arange(0,.1,.0001)#gives me a set of y values to plot a vertical line

    hist(y,bins=160,normed=True)#plots a histagram of the y data
    n,bins,junk=hist(y,bins=160,normed=True,label="histogram of generated colition data for "+atom1+" and "+atom2+" collision")#gives me the data for the histagram plot
    csum_normed=cumsum(n)/sum(n)#by using a method similar to how I created the nucleons I am looking for 10 percent and 20 percent of the colitions on the high end
    #note: I will put a better disciption of this method in the bellow markdown cell
    percent_10=bins[csum_normed.searchsorted(.9)]#finds where 10 percent of the colitions ocure on the high end
    percent_20=bins[csum_normed.searchsorted(.8)]#finds where 20 percent of the colitions ocure
    percent_vert_10=percent_10*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    percent_vert_20=percent_20*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    plot(percent_vert_10,yforploting,label="10percent")#plots vertical line at 10 percent
    plot(percent_vert_20,yforploting,label="20percent")#plots vertical line at 20 percent
    legend(loc="best")
    xlabel("number of collisions")
    ylabel("percent chance of colition # ocuring")
    subplot(2,1,1)#creats the secound plot
    xlabel("distance between atoms in fm")
    ylabel("number of nucleons that colide")
    plot(x,y,"r*",label="generated collision data for "+atom1+" and "+atom2+" collision")#plots the atom number of clitions vs distance
    legend(loc="best")
    xlim(0,10)#sets the x limutes
```

description of cumsum method:
A cumsum is a continous sum over the values. Here it allowed me to find the 10 and 20 percent ranges by summing the number of events and finding where 10/20 percent of the events ocure with high counts.

```python
def collision_cumsum_plot(atom1,atom2,Nc,energy):
    """given atom 1 atom 2 number of runs and energy in GeV/c returns plots of collision data with cumsum plot"""
    import numpy as np
    x=zeros(Nc)#intialize two arays the with leghts equal to the number of atoms you want to sample
    y=zeros(Nc)
    d=find_interacting_distance(energy)#gives radius of the proton at this energy
    for b in range(0,Nc):#the loop goes through the counting done in ploting_2 the number of times indicated by Nc
        count=0#sets count to 0
        atom1_dis=random_r(atom1)#gets the data on the nucleons
        atom2_dis=random_r(atom2)
        atom1_distance_cordinates=((atom1_dis[0]*cos(atom1_dis[1])),(atom1_dis[0]*sin(atom1_dis[1])))#converts to x,y cordinates
        atom2_distance_cordinates=(((atom2_dis[0]*cos(atom2_dis[1]))+atom1_dis[4]+atom2_dis[4]),(atom2_dis[0]*sin(atom2_dis[1])))
        #seperate x,y data
        x_atom1=atom1_distance_cordinates[0]
        y_atom1=atom1_distance_cordinates[1]
        x_atom2=atom2_distance_cordinates[0]
        y_atom2=atom2_distance_cordinates[1]
        count=0
        #the bellow for loops determine if the nuclei colide
        for i in range(0,len(x_atom1)):
            for c in range(0,len(x_atom2)):
                distance=((x_atom1[i]-x_atom2[c])**2+(y_atom1[i]-y_atom2[c])**2)**.5#finds distacne between each nucleon
                if distance<2*d:
                    count+=1

        x[b]=atom1_dis[4]+atom2_dis[4]#stors the distance between each atom in array
        y[b]=count#adds the count to an array
    figure(figsize=(15,10))#set figure size

    #same ploting as in plotting_2
    yforploting=np.arange(0,.1,.0001)

    hist(y,bins=160,normed=True)
    n,bins,junk=hist(y,bins=160,normed=True,label="histogram of collision data for "+atom1+" and "+atom2+" collision")
    csum_normed=cumsum(n)/sum(n)
    percent_10=bins[csum_normed.searchsorted(.9)]
    percent_20=bins[csum_normed.searchsorted(.8)]
    percent_vert_10=percent_10*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    percent_vert_20=percent_20*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    plot(percent_vert_10,yforploting,label="10percent")#plots vertical line at 10 percent
    plot(percent_vert_20,yforploting,label="20percent")#plots vertical line at 20 percent
    plot(csum_normed,label="cumsum curve for "+atom1+" and "+atom2+"collision")#plot the normalized cumsum
    legend(loc="upper right")
    xlabel("Number of collisions")
    ylabel("percent chance of number occuring")

```

```python
def Participating_data_plot(atom1,atom2,Nc,energy):#need the two atoms the number of samples you want Nc and the energy you want the colitions to take place at
    """given atom 1 atom 2 number of runs and energy in GeV/c returns plots of participation data"""
    import numpy as np
    x=zeros(Nc)#intialize two arays the with leghts equal to the number of atoms you want to sample
    y=zeros(Nc)
    d=find_interacting_distance(energy)#gives radius of the proton at this energy
    for b in range(0,Nc):#the loop goes through the counting done in ploting_2 the number of times indicated by Nc
        count=0#sets count to 0
        atom1_dis=random_r(atom1)#gets the data on the nucleons
        atom2_dis=random_r(atom2)
        atom1_distance_cordinates=((atom1_dis[0]*cos(atom1_dis[1])),(atom1_dis[0]*sin(atom1_dis[1])))#converts to x,y cordinates
        atom2_distance_cordinates=(((atom2_dis[0]*cos(atom2_dis[1]))+atom1_dis[4]+atom2_dis[4]),(atom2_dis[0]*sin(atom2_dis[1])))
        #seperate x,y data
        x_atom1=atom1_distance_cordinates[0]
        y_atom1=atom1_distance_cordinates[1]
        x_atom2=atom2_distance_cordinates[0]
        y_atom2=atom2_distance_cordinates[1]
        count=0
        #the bellow for loops determine if the nuclei colide
        count_participants_from_atom1=0
        count_participants_from_atom2=0
        count_participants=0
        for i in range(0,len(x_atom1)):#loops through all of atom 1 nuclei
            counter_par=0
            for c in range(0,len(x_atom2)):#loops through all of atom 2 nuclei
                distance=((x_atom1[i]-x_atom2[c])**2+(y_atom1[i]-y_atom2[c])**2)**.5#finds the distance between atom 1 nuclei and atom 2 nuclei
                if distance<2*d:#checks to see if the distance is less than 2 times the radius of the proton(meaning the nucleon colide)
                    if counter_par==0:#check to see if this is the nucleid's first time to this point
                        count_participants+=1#it is a perticipant
                        count_participants_from_atom1+=1#it is a perticapant from atom 1
                        counter_par=1#says this nuclei can not get back into loop
        for i in range(0,len(x_atom2)):#loops through all of atom 2 nuclei
            counter_par=0
            for c in range(0,len(x_atom1)):#loops through all of atom 1 nuclei
                distance=((x_atom1[c]-x_atom2[i])**2+(y_atom1[c]-y_atom2[i])**2)**.5#finds the distance between atom 1 nuclei and atom 2 nuclei
                if distance<2*d:#checks to see if the distance is less than 2 times the radius of the proton(meaning the nucleon colide)
                    if counter_par==0:#check to see if first time to this point
                        count_participants+=1#adds 1 if perticipates
                        count_participants_from_atom2+=1#adds 1 if perticipates
                        counter_par=1
    

        x[b]=atom1_dis[4]+atom2_dis[4]#stors the distance between each atom in array
        y[b]=count_participants#adds the count to an array
    figure(figsize=(15,15))#set the figure size
    subplot(2,1,2)#plots the lower graph in the subplot
    
    yforploting=np.arange(0,.5,.01)#gives me a set of y values to plot a vertical line
    numbins=atom1_dis[5]+atom2_dis[5]
    hist(y,bins=numbins,normed=True)#plots a histagram of the y data
    n,bins,junk=hist(y,bins=numbins,normed=True,label="histogram of nuclei participating data for "+atom1+" and "+atom2+"collision")#gives me the data for the histagram plot
    csum_normed=cumsum(n)/sum(n)#by using a method similar to how I created the nucleons I am looking for 10 percent and 20 percent of the colitions on the high end
    #note: I will put a better disciption of this method in the bellow markdown cell
    percent_10=bins[csum_normed.searchsorted(.9)]#finds where 10 percent of the colitions ocure on the high end
    percent_20=bins[csum_normed.searchsorted(.8)]#finds where 20 percent of the colitions ocure
    percent_vert_10=percent_10*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    percent_vert_20=percent_20*ones(len(yforploting))#gives me an array of all the same value to plot a vertical line
    plot(percent_vert_10,yforploting,label="10 percent")#plots vertical line at 10 percent
    plot(percent_vert_20,yforploting,label="20 percent")#plots vertical line at 20 percent
    legend(loc="best")
    xlabel("number of participants")
    ylabel("percent chance of colition # ocuring")
    subplot(2,1,1)#creats the secound plot
    xlabel("distance between atoms in fm")
    ylabel("number of nucleons that participate")
    plot(x,y,"r*",label="generated colition data for "+atom1+" and "+atom2+"collision")#plots the atom number of clitions vs distance
    legend(loc="best")
    xlim(0,10)#sets the x limutes
```
