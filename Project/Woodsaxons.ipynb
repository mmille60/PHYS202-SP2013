```python
def Woodsaxson(atom):
    p_r=lambda r: p0*(1+(w*r**2)/R0)/(1+exp((r-R0)/a))
    p0=1
    if atom=='O' or atom ==16:
        w=-0.051
        R0=2.608
        a=.513
    elif atom=='C' or atom==12:
        w=0.
        R0=2.47
        a=0.
    elif atom=='Al' or atom==27:
        R0=3.07
        a=.519
        w=0.
    elif atom=='S' or atom==32:
        R0=3.458
        a=.61
        w=0
    elif atom=="Ca" or atom==40:
        w=-0.161
        R0=3.76
        a=.586
    elif atom=='Ni' or atom==58:
        R0=4.309
        a=.516
        w=-0.1308
    elif atom=='Cu' or atom==63:
        R0=4.2
        a=.596
        w=0
    elif atom=='W' or atom==186:
        R0=6.51
        a=.535
        w=0
    elif atom=='Au' or atom==197:
        R0=6.38
        a=.535
        w=0
    elif atom=='Pb' or atom==208:
        R0=6.68
        a=.546
        w=0
    elif atom=='U' or atom==238:
        R0=6.68
        a=.6
        w=0
    r=np.arange(0,2*R0,.01)
    return p_r(r)
```

```python
rO=np.arange(0,2*2.608,.01)
plot(rO,Woodsaxson("O"),label="oxygen dencity plot")
rCu=np.arange(0,2*4.2,.01)
plot(rCu,Woodsaxson("Cu"),label="copper dencity plot")
rPb=np.arange(0,2*6.68,.01)
plot(rPb,Woodsaxson("Pb"),label="lead dencity plot")
legend(loc="best")
xlabel('radius(fm)')
ylabel('dencity')
show()
```
