```python
def random_r(number,atom):
    N=number
    z=numpy.random.random_sample(N);
    F1=cumsum(Woodsaxson(atom))
    a=sum(Woodsaxson(atom))
    F1=F1/a
    T=F1.searchsorted(z)
    r=Woodsaxson_for_R(atom)
    new=r[T]
    return new
```

```python
def random_r_and_theta(number,atom):
    import random
    import numpy as np
    N=number
    z=numpy.random.random_sample(N);
    F1=cumsum(Woodsaxson(atom))
    a=sum(Woodsaxson(atom))
    F1=F1/a
    T=F1.searchsorted(z)
    r=Woodsaxson_for_R(atom)
    new=r[T]
    theta=zeros(len(new))
    t=0
    for i in range(0,len(new)):
        theta[t]=random.uniform(0,2*pi)
        t+=1
    return (theta,new)
```
